```python
n = int(input("Введите n: "))
a = [int(input("Введите число:")) for i in range (n)]
ch = 0
ins = 0
ot = 0
for i in range (n):
  if a[i]%2 == 0:
    ch += 1
  if a[i] > -256 and a[i] < 1024:
    ins += 1
  if a[i] < 0:
    ot +=1
print ("количество четных чисел=", ch)
print ("количество отрицательных чисел=", ot)
print ("количество чисел промежутке [-256 ; 1024]=", ins)
for i in range (n):
  for j in range(i+1, n):
    if a[i] == a[j]:
      print("есть одинаковые элементы")
      exit
```
